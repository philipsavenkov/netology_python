# Task1

Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность 
определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

Результат классификации запишите в отдельный столбец region.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('keywords.csv')
```

```python
geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}
```

```python
geo_data_inv = {}

for key in geo_data:
    for value in geo_data[key]:
        geo_data_inv[value] = key
```

```python
def process(keyword):
    for k in geo_data_inv.keys():
        if k in keyword:
            return geo_data_inv[k]
    return 'undefined'
```

```python
data['region'] = data['keyword'].apply(lambda x: process(x))
```

# Task2

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
    - оценка 2 и меньше - низкий рейтинг
    - оценка 4 и меньше - средний рейтинг
    - оценка 4.5 и 5 - высокий рейтинг
Результат классификации запишите в столбец class

```python
df_movies = pd.read_csv('ml-latest-small/movies.csv')
#df_movies.head()
```

```python
df_ratings = pd.read_csv('ml-latest-small/ratings.csv')
#df_ratings.head()
```

```python
df_merged = pd.merge(df_movies, df_ratings, on='movieId')
df_merged.head()
```

```python
#df_merged.rating.unique()
```

```python
def class_(row):
    if row['rating'] <= 2:
        rate = 'Низкий рейтинг'
    elif row['rating'] <= 4:
        rate = 'Средний рейтинг'
    else:
        rate = 'Высокий рейтинг'
    return rate
```

```python
df_merged['class'] = df_merged.apply(class_, axis=1)
```

```python
def year(row):
    for i in range(0, len(years)):
        if str(years[i]) in row['title']:
            return str(years[i])
        else:
            i += 1
    return '1900'
```
